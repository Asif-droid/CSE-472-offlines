```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_vectors(vectors, colors, scale=60):
    V = np.array([[0, 1], [1, 0]])  # ihat and jhat
    colors = ['k', 'k'] + colors
    V = np.vstack([V, np.hstack(vectors).T])
    print(V)

    plt.figure(figsize=(5, 5))

    plt.grid()

    origin = np.zeros(V.T.shape)
    plt.quiver(*origin, V[:,0], V[:,1], color=colors, scale=1, units='xy')

    lim = V.max()*2

    plt.xlim(-lim, lim)
    plt.ylim(-lim, lim)

    plt.show()
```

```python
def generate_invertible_matrix(n):
    while True:
        # Create a random n x n matrix with entries
        # random_matrix = np.random.randint(np.iinfo(np.int32).min, np.iinfo(np.int32).max, size=(n, n))
        random_matrix = np.random.randint(-100, 100, size=(n, n))

        # Check if the matrix is invertible
        if np.linalg.matrix_rank(random_matrix) == n:

            return random_matrix


# A=generate_invertible_matrix(2)
# A
```

```python
from typing import AbstractSet
n=input()

A=generate_invertible_matrix(int(n))
A
```

```python
eigenvalues, eigenvectors = np.linalg.eig(A)

print(eigenvalues)
print(eigenvectors)
```

```python
A_reconstructed = eigenvectors @ np.diag(eigenvalues) @ np.linalg.inv(eigenvectors)

are_matrices_equal = np.allclose(A_reconstructed, A)

if are_matrices_equal:
    print("The matrices are approximately equal.")
else:
    print("The matrices are not equal.")
```

```python
def generate_invertible_symmetric_matrix(n):
    """
    Generate a random invertible symmetric matrix of size n x n using NumPy.
    """
    while True:
       # Generate a random lower triangular matrix with integers
        lower_triangular = np.tril(np.random.randint(-100, 100, size=(n, n)), k=-1)

        # Create a symmetric matrix by filling the upper triangular part
        symmetric_matrix = lower_triangular + lower_triangular.T

        # Set the diagonal to random integers
        np.fill_diagonal(symmetric_matrix, np.random.randint(-100, 100, size=n))

        # Check if the matrix is invertible by measuring rank
        if np.linalg.matrix_rank(symmetric_matrix) == n:
            return symmetric_matrix.astype(int)

# AS = generate_invertible_symmetric_matrix(5)
# AS
```

```python
from typing import AbstractSet
n2=input()

As=generate_invertible_symmetric_matrix(int(n2))
As
```

```python
eigenvalues, eigenvectors = np.linalg.eig(As)

print(eigenvalues)
print(eigenvectors)
```

```python
As_reconstructed = eigenvectors @ np.diag(eigenvalues) @ np.linalg.inv(eigenvectors)

are_matrices_equal = np.allclose(As_reconstructed, As)

if are_matrices_equal:
    print("The matrices are approximately equal.")
else:
    print("The matrices are not equal.")
```
